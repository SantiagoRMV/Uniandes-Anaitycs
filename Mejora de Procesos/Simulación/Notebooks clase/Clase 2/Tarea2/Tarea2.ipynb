# Analisis de Bondad y Ajuste con Python

```python
import pandas as pd
df = pd.read_csv('mmps_clase2_datos.csv')
```

### Analisis general

```python
import numpy as np
from scipy import stats
from distfit import distfit

dfit = distfit(todf=True)
dfit.fit_transform(df['TiempoEntreArribos'])
dfit.plot()
```

```python
dfit.plot_summary()
```

```python
def pruebas_de_bondad(dfit, data, iterar):  #Iterar para ver si el titulo va por franja o general
    # Obtener la mejor distribución y sus parámetros
    best_distr_name = dfit.model['name']
    best_distr_params = dfit.model['params']

    # Realizar la prueba de Kolmogorov-Smirnov
    # Se usa la distribución y parámetros obtenidos para generar la CDF
    distr_function = getattr(stats, best_distr_name)
    ks_stat, ks_p_value = stats.kstest(data, best_distr_name, args=best_distr_params)

    if iterar:
        print("--------------------------------------------------")
        print(f"Pruebas de bondad: Franja horaria {i}")
        print("--------------------------------------------------")
    else:
        print("--------------------------------------------------")
        print(f"Pruebas de bondad General")
        print("--------------------------------------------------")

    print(f"Distribucion {best_distr_name}")
    print("--------------------------------------------------")
    print(f"KS test statistic: {ks_stat}")
    print(f"KS test p-value: {ks_p_value}")
    print("--------------------------------------------------")

    # Calcular los cuantiles para dividir los datos en intervalos equiprobables
    num_bins = 10  
    quantiles = np.linspace(0, 1, num_bins+1)  # los cuantiles 
    bin_edges = distr_function.ppf(quantiles, *best_distr_params)  # calcular los cuantiles de la distribución

    # Calcular las frecuencias observadas y esperadas
    observed_freq, _ = np.histogram(data, bins=bin_edges)  
    total_observed = observed_freq.sum()
    expected_freq = np.full_like(observed_freq, fill_value=total_observed / num_bins)

    # Ajustar las frecuencias esperadas si no suman exactamente lo mismo que las observadas
    difference = total_observed - expected_freq.sum()
    if difference != 0:
        expected_freq[-1] += difference  # ajustar la última frecuencia esperada

    
    chi_stat, chi_p_value = stats.chisquare(f_obs=observed_freq, f_exp=expected_freq)
    print("--------------------------------------------------")
    print(f"Chi-square test statistic: {chi_stat}")
    print(f"Chi-square test p-value: {chi_p_value}")
    print("--------------------------------------------------")

data = df['TiempoEntreArribos']
pruebas_de_bondad(dfit, data, False)
```

Haremos las pruebas de bondad solo para datos entre IQR

```python
from scipy.stats import t
# Parámetros
dfr = 1.57844  # grados de libertad 
loc = 5.888  # media de la distribución
scale = 1.638  # desviación estándar de la distribución

# Cuartiles
q1 = t.ppf(0.25, dfr, loc, scale)  # primer cuartil
q3 = t.ppf(0.75, dfr, loc, scale)  # tercer cuartil

# Rango intercuartílico
iqr = q3 - q1

limite_inferior = q3  - 1.5*iqr
limite_superior = q1  + 1.5*iqr

data = df[(df['TiempoEntreArribos']<= limite_superior) & (df['TiempoEntreArribos'] >= limite_inferior)]['TiempoEntreArribos']
pruebas_de_bondad(dfit, data, False)
```

# Generación de números aleatorios

## 1. Distribución Normal

### Iniciamos generando números aleatorios de acuerdo con una distribución normal.

```python
import numpy as np
from scipy.stats import norm
n=1000
loc=3
scale=0.5
vals_normal = norm.rvs(loc=loc, scale=scale, size=n)
```

### Ahora calculamos el histograma asociado a los datos y lo graficamos junto con la densidad de la normal.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
# histograma
count, bins, ignored = plt.hist(x=vals_normal, bins=30, density=True)

# densidad
nobs = 10000
xmin = min(vals_normal)
xmax = max(vals_normal)
step = (xmax-xmin)/nobs
x= np.arange(xmin,xmax,step)
plt.plot(x, norm.pdf(x,loc=3,scale=0.5))

plt.title('Histograma y densidad')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

### Para la misma distribución podemos calcular los primeros 4 momentos.

```python
mean, var, skew, kurt = norm.stats(loc=loc,scale=scale, moments='mvsk')
print("Media: {}".format(mean))
print("Varianza: {}".format(var))
print("Asimetría: {}".format(skew))
print("Curtosis: {}".format(kurt))
```

### Ahora tomamos los datos generados y los usamos para ajustar los parámetros de una variable normal. Calculamos los momentos de la variable ajustada y mostramos la densidad de ésta contra el histograma de los datos. 

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
# ajuste de datos a una normal - obtener parámetros
floc, fscale = norm.fit(vals_normal)
print("Parámetros de la distribución normal ajustada")
print("Localización: {}".format(floc))
print("Escala: {}".format(fscale))


# variable ajustada 
fitted_var = norm.freeze(loc=floc, scale=fscale)
# momentos de la variable ajustada
mean, var, skew, kurt = fitted_var.stats(moments='mvsk')
print("\nMomentos de la distribución normal ajustada")
print("Media: {}".format(mean))
print("Varianza: {}".format(var))
print("Asimetría: {}".format(skew))
print("Curtosis: {}".format(kurt))


# histograma
count, bins, ignored = plt.hist(x=vals_normal, bins=30, density=True)

# densidad
nobs = 10000
xmin = min(vals_normal)
xmax = max(vals_normal)
step = (xmax-xmin)/nobs
x= np.arange(xmin,xmax,step)
plt.plot(x, fitted_var.pdf(x))

plt.title('Histograma y densidad')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

## Distribución Gamma

Repetimos el ejercicio anterior con una distribución Gamma.

```python
import numpy as np
from scipy.stats import gamma
n=1000
a = 5
loc = 1
vals_gamma = gamma.rvs(a=a, loc=loc, size=n)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import gamma
# histograma
count, bins, ignored = plt.hist(x=vals_gamma, bins=30, density=True)

# densidad
nobs = 10000
xmin = min(vals_gamma)
xmax = max(vals_gamma)
step = (xmax-xmin)/nobs
x= np.arange(xmin,xmax,step)
plt.plot(x, gamma.pdf(x, a=a,loc=loc))

plt.title('Histograma y densidad')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

```python
mean, var, skew, kurt = gamma.stats(a=a,loc=loc, moments='mvsk')
print("Media: {}".format(mean))
print("Varianza: {}".format(var))
print("Asimetría: {}".format(skew))
print("Curtosis: {}".format(kurt))
```

```python
import matplotlib.pyplot as plt
from scipy.stats import gamma
# ajuste de datos a una gamma - obtener parámetros
fa, floc, fscale = gamma.fit(vals_gamma)
print("Parámetros de la distribución gamma ajustada")
print("a: {}".format(fa))
print("Localización: {}".format(floc))
print("Escala: {}".format(fscale))

# variable ajustada 
fitted_var = gamma.freeze(a=fa,loc=floc,scale=fscale)
# momentos de la variable ajustada
mean, var, skew, kurt = fitted_var.stats(moments='mvsk')
print("\nMomentos de la distribución gamma ajustada")
print("Media: {}".format(mean))
print("Varianza: {}".format(var))
print("Asimetría: {}".format(skew))
print("Curtosis: {}".format(kurt))


# histograma
count, bins, ignored = plt.hist(x=vals_gamma, bins=30, density=True)

# densidad
nobs = 10000
xmin = min(vals_gamma)
xmax = max(vals_gamma)
step = (xmax-xmin)/nobs
x= np.arange(xmin,xmax,step)
plt.plot(x, fitted_var.pdf(x))

plt.title('Histograma y densidad')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

La distribución Gamma se conforma de los siguientes parámetros:
1. a: Determina la forma de la distribución y controla la asimetría

> El valor de α determina la forma de la distribución. Un valor más alto de α hará que la distribución gamma sea más simétrica y se concentre más alrededor de su media. Un valor más bajo de α hará que la distribución sea más asimétrica, con una cola derecha más larga.

> Si a = integer la función se reduce a una distribución Erlang. 
Si a = 1, toma la forma de una distribución exponencial.

* a -> 5 a 1

```python
mean, var, skew, kurt = gamma.stats(a=a,loc=loc, moments='mvsk')
print("Media: {}".format(mean))
print("Varianza: {}".format(var))
print("Asimetría: {}".format(skew))
print("Curtosis: {}".format(kurt))
```

```python
# ajuste de datos a una gamma - obtener parámetros
fa, floc, fscale = gamma.fit(vals_gamma_2)
print("Parámetros de la distribución gamma ajustada")
print("a: {}".format(fa))
print("Localización: {}".format(floc))
print("Escala: {}".format(fscale))

# variable ajustada 
fitted_var = gamma.freeze(a=fa,loc=floc,scale=fscale)
# momentos de la variable ajustada
mean, var, skew, kurt = fitted_var.stats(moments='mvsk')
print("\nMomentos de la distribución gamma ajustada")
print("Media: {}".format(mean))
print("Varianza: {}".format(var))
print("Asimetría: {}".format(skew))
print("Curtosis: {}".format(kurt))


# histograma
count, bins, ignored = plt.hist(x=vals_gamma_2, bins=30, density=True)

# densidad
nobs = 10000
xmin = min(vals_gamma_2)
xmax = max(vals_gamma_2)
step = (xmax-xmin)/nobs
x= np.arange(xmin,xmax,step)
plt.plot(x, fitted_var.pdf(x))

plt.title('Histograma y densidad - modificada')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

Las distribuciones gaussianas inversas tienen muchas similitudes con las distribuciones gaussianas estándar (normales), lo que conduce a aplicaciones en estadística inferencial.

La distribución gaussiana inversa utiliza los siguientes parámetros.

* μ: Parámetro de escala,	μ>0, especificado como un valor de escalar positivo.
* λ: Parámetro de forma,	λ>0, especificado como valor de escalar positivo.


$$f(x; \mu, \lambda) = \sqrt{\frac{\lambda}{2\pi x^3}} \exp\left(-\frac{\lambda (x - \mu)^2}{2\mu^2 x}\right)
$$
