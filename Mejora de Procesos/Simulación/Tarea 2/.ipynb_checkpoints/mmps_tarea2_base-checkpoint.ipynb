## Tarea 2 - Generación de números aleatorios

## 1. Distribución Normal

### Iniciamos generando números aleatorios de acuerdo con una distribución normal.

```python
import numpy as np
from scipy.stats import norm
n=1000
loc=3
scale=0.5
vals_normal = norm.rvs(loc=loc, scale=scale, size=n)
```

### Ahora calculamos el histograma asociado a los datos y lo graficamos junto con la densidad de la normal.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
# histograma
count, bins, ignored = plt.hist(x=vals_normal, bins=30, density=True)

# densidad
nobs = 10000
xmin = min(vals_normal)
xmax = max(vals_normal)
step = (xmax-xmin)/nobs
x= np.arange(xmin,xmax,step)
plt.plot(x, norm.pdf(x,loc=3,scale=0.5))

plt.title('Histograma y densidad')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

### Para la misma distribución podemos calcular los primeros 4 momentos.

```python
mean, var, skew, kurt = norm.stats(loc=loc,scale=scale, moments='mvsk')
print("Media: {}".format(mean))
print("Varianza: {}".format(var))
print("Asimetría: {}".format(skew))
print("Curtosis: {}".format(kurt))
```

### Ahora tomamos los datos generados y los usamos para ajustar los parámetros de una variable normal. Calculamos los momentos de la variable ajustada y mostramos la densidad de ésta contra el histograma de los datos. 

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
# ajuste de datos a una normal - obtener parámetros
floc, fscale = norm.fit(vals_normal)
print("Parámetros de la distribución normal ajustada")
print("Localización: {}".format(floc))
print("Escala: {}".format(fscale))


# variable ajustada 
fitted_var = norm.freeze(loc=floc, scale=fscale)
# momentos de la variable ajustada
mean, var, skew, kurt = fitted_var.stats(moments='mvsk')
print("\nMomentos de la distribución normal ajustada")
print("Media: {}".format(mean))
print("Varianza: {}".format(var))
print("Asimetría: {}".format(skew))
print("Curtosis: {}".format(kurt))


# histograma
count, bins, ignored = plt.hist(x=vals_normal, bins=30, density=True)

# densidad
nobs = 10000
xmin = min(vals_normal)
xmax = max(vals_normal)
step = (xmax-xmin)/nobs
x= np.arange(xmin,xmax,step)
plt.plot(x, fitted_var.pdf(x))

plt.title('Histograma y densidad')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

### Modificación de parámetros - Distribución Gamma

La distribución Gamma se conforma de los siguientes parámetros:
1. a: Determina la forma de la distribución y controla la asimetría

> El valor de α determina la forma de la distribución. Un valor más alto de α hará que la distribución gamma sea más simétrica y se concentre más alrededor de su media. Un valor más bajo de α hará que la distribución sea más asimétrica, con una cola derecha más larga.

> Si a = integer la función se reduce a una distribución Erlang. 
Si a = 1, toma la forma de una distribución exponencial.


Se realizo la generación de numeros aleatorios para la distribución Gamma con cambiando los parámetros:
* a -> 5 a 1

```python
n=1000
a = 1 #Cambio de 5 a 1
loc = 1 #Se mantiene
vals_gamma_2 = gamma.rvs(a=a, loc=loc, size=n)

# histograma
count, bins, ignored = plt.hist(x=vals_gamma_2, bins=30, density=True)

# densidad
nobs = 10000
xmin = min(vals_gamma_2)
xmax = max(vals_gamma_2)
step = (xmax-xmin)/nobs
x= np.arange(xmin,xmax,step)
plt.plot(x, gamma.pdf(x, a=a,loc=loc))

plt.title('Histograma y densidad - modificado')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

```python
mean, var, skew, kurt = gamma.stats(a=a,loc=loc, moments='mvsk')
print("Media: {}".format(mean))
print("Varianza: {}".format(var))
print("Asimetría: {}".format(skew))
print("Curtosis: {}".format(kurt))
```

```python
# ajuste de datos a una gamma - obtener parámetros
fa, floc, fscale = gamma.fit(vals_gamma_2)
print("Parámetros de la distribución gamma ajustada")
print("a: {}".format(fa))
print("Localización: {}".format(floc))
print("Escala: {}".format(fscale))

# variable ajustada 
fitted_var = gamma.freeze(a=fa,loc=floc,scale=fscale)
# momentos de la variable ajustada
mean, var, skew, kurt = fitted_var.stats(moments='mvsk')
print("\nMomentos de la distribución gamma ajustada")
print("Media: {}".format(mean))
print("Varianza: {}".format(var))
print("Asimetría: {}".format(skew))
print("Curtosis: {}".format(kurt))


# histograma
count, bins, ignored = plt.hist(x=vals_gamma_2, bins=30, density=True)

# densidad
nobs = 10000
xmin = min(vals_gamma_2)
xmax = max(vals_gamma_2)
step = (xmax-xmin)/nobs
x= np.arange(xmin,xmax,step)
plt.plot(x, fitted_var.pdf(x))

plt.title('Histograma y densidad - modificada')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

Cuando se modifica el valor de a para que sea 1, la gráfica toma la forma de una distribución exponencial. Al disminuir el valor de a, la asimetría es positiva con el mayor volumen de datos en el lado izquierdo y una cola larga al derecho.

## Distribución Gaussiana Inversa (distribución propuesta)

La distribución gaussiana inversa, también conocida como distribución de Wald, se utiliza para modelar datos no negativos con sesgo positivo. Las distribuciones gaussianas inversas tienen muchas similitudes con las distribuciones gaussianas estándar (normales), lo que conduce a aplicaciones en estadística inferencial.

La distribución gaussiana inversa utiliza los siguientes parámetros.

* μ: Parámetro de escala,	μ>0, especificado como un valor de escalar positivo.
* λ: Parámetro de forma,	λ>0, especificado como valor de escalar positivo.


$$f(x; \mu, \lambda) = \sqrt{\frac{\lambda}{2\pi x^3}} \exp\left(-\frac{\lambda (x - \mu)^2}{2\mu^2 x}\right)
$$

```python
import math
import numpy as np
from scipy.stats import invgauss
n=1000
mu=1
loc = 1
vals_invgauss = invgauss.rvs(mu=mu, loc=loc, size=n)

# histograma
count, bins, ignored = plt.hist(x=vals_invgauss, bins=30, density=True)

# densidad
nobs = 10000
xmin = min(vals_invgauss)
xmax = max(vals_invgauss)
step = (xmax-xmin)/nobs
x= np.arange(xmin,xmax,step)
plt.plot(x, weibull_min.pdf(x, c=c, loc=loc))


plt.title('Histograma y densidad')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

### Modificación Parámetros - Distribución Gauss Inversa

La distribución Gaussiana inversa se conforma de los siguientes parámetros:
1. μ (mu): Determina la media de la distribución. Controla la ubicación de la distribución en el eje x.

> Esta distribución es especialmente útil para modelar variables que se distribuyen según una cola larga hacia valores positivos.

Se realizo la generación de numeros aleatorios para la distribución Gauss Inversa con cambiando los parámetros:
*  μ (mu): -> 1 a 0.5

```python
n=1000
mu=0.5
loc = 1
vals_invgauss_2 = invgauss.rvs(mu=mu, loc=loc, size=n)

# histograma
count, bins, ignored = plt.hist(x=vals_invgauss_2, bins=30, density=True)

# densidad
nobs = 10000
xmin = min(vals_invgauss_2)
xmax = max(vals_invgauss_2)
step = (xmax-xmin)/nobs
x= np.arange(xmin,xmax,step)
plt.plot(x, weibull_min.pdf(x, c=c, loc=loc))


plt.title('Histograma y densidad')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

```python
mean, var, skew, kurt = invgauss.stats(mu=mu, loc=loc, moments='mvsk')
print("Media: {}".format(mean))
print("Varianza: {}".format(var))
print("Asimetría: {}".format(skew))
print("Curtosis: {}".format(kurt))
```

```python
# ajuste de datos a una Invgauss - obtener parámetros
fm, floc, fscale = invgauss.fit(vals_invgauss_2)
print("Parámetros de la distribución Gauss Inversa ajustada")
print("mu: {}".format(fm))
print("Localización: {}".format(floc))
print("Escala: {}".format(fscale))

# variable ajustada 
fitted_var = invgauss.freeze(mu=fc, loc=floc,scale=fscale)
# momentos de la variable ajustada
mean, var, skew, kurt = fitted_var.stats(moments='mvsk')
print("\nMomentos de la distribución Gauss Inversa ajustada")
print("Media: {}".format(mean))
print("Varianza: {}".format(var))
print("Asimetría: {}".format(skew))
print("Curtosis: {}".format(kurt))


# histograma
count, bins, ignored = plt.hist(x=vals_invgauss_2, bins=30, density=True)

# densidad
nobs = 10000
xmin = min(vals_invgauss_2)
xmax = max(vals_invgauss_2)
step = (xmax-xmin)/nobs
x= np.arange(xmin,xmax,step)
plt.plot(x, fitted_var.pdf(x))

plt.title('Histograma y densidad - modificado')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

Cuando se disminuye el parámetro de μ, se ve como se reduce la cola de 10 a 4 y la dispersión es menor a la original. Antes de relizar el ajuste de los datos, se puede ver como la linea de referencia de la distribución se comienza a aplanar, comportamiento que ocurre a medida que μ se acerca a 0.
