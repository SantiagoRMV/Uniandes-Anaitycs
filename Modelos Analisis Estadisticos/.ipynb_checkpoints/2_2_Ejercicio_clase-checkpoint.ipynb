```python
from scipy.stats import norm
```

**10.** Las puntuaciones de satisfacci칩n de los empleados en una empresa est치n normalmente distribuidas con una media de 70 y una desviaci칩n est치ndar de 15. 쮺u치l es la probabilidad de que un empleado seleccionado al azar tenga una puntuaci칩n mayor que 85?

```python
mean_10 = 70
std_10 = 15
Puntaje = 85
z = (Puntaje - mean_10) / std_10
Prob_acumulada = norm.cdf(z)
Prob = 1 - Prob_acumulada
print(Prob)
```

**11.** Un restaurante estudia el tiempo que los clientes pasan cenando y encuentra que sigue una distribuci칩n normal con una media de 45 minutos y una desviaci칩n est치ndar de 10 minutos. 쮺u치l es el tiempo que supera el 90% de los clientes?

```python
#Usar norm.cdf = Cumulative Distribution Function
mean_11 = 45
std_11 = 10
prob_acum_11 = 0.9

prob_x = 1 - prob_acum_11
z = norm.cdf(prob_x)
x = (z*std_11) + mean_11
print(x)
```

**12.** Las ventas de una empresa en particular se distribuyen uniformemente entre 50 y 150 unidades diarias. Si un d칤a venden 100 unidades, 쯤u칠 porcentaje de d칤as vender치n m치s que eso?

```python
from scipy.stats import uniform
```

```python
a = 50
b = 150
ventas = 100
x = uniform.cdf(ventas, loc=a, scale=b-a)
print(x)
```

**13.** Si las llamadas al centro de atenci칩n al cliente tienen una media de duraci칩n de 6 minutos con una desviaci칩n est치ndar de 2 minutos, y asumimos que las duraciones siguen una distribuci칩n normal, 쯖u치l es la probabilidad de que una llamada dure menos de 4 minutos?

```python
mean_13 = 6
std_13 = 2
duracion = 4
z = (mean_13 - duracion)/std_13
prob_x_13 = norm.cdf(z)
print(prob_x_13)
```

**15.** Un software de gesti칩n de proyectos mide el tiempo que un empleado tarda en completar una tarea. Si se sabe que el tiempo se distribuye normalmente con una media de 3 horas y una desviaci칩n est치ndar de 30 minutos, 쯖u치l es la probabilidad de que una tarea se complete en menos de 2.5 horas?

```python
mean_15 = 3
std_15 = 0.5
tiempo = 2.5
z = (mean_15 - tiempo) / std_15
prob_x_15 = norm.cdf(z)
print(prob_x_15)
```
