```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = '/Users/santiagoromero/Documents/Maestria/Modelos Analisis Estadisticos/Tareas/Tarea 1 2/Data_Normalizada.csv'
data_frame = pd.read_csv(data, sep=';', low_memory=False)
data_frame
```

```python
data_frame['Normal_PUNT_MATEMATICAS'] = data_frame['Normal_PUNT_MATEMATICAS'].str.replace(',', '.', regex=True).astype(float) 
data_frame['Normal_PUNT_C_NATURALES'] = data_frame['Normal_PUNT_C_NATURALES'].str.replace(',', '.', regex=True).astype(float) 
data_frame['Normal_PUNT_SOCIALES_CIUDADANAS'] = data_frame['Normal_PUNT_SOCIALES_CIUDADANAS'].str.replace(',', '.', regex=True).astype(float) 
data_frame['Normal_PUNT_INGLES'] = data_frame['Normal_PUNT_INGLES'].str.replace(',', '.', regex=True).astype(float) 
data_frame['Normal_PUNT_GLOBAL'] = data_frame['Normal_PUNT_GLOBAL'].str.replace(',', '.', regex=True).astype(float) 
```

```python
data_frame.info()
```

```python
#Obteniendo las variables continuas
variables_continuas = []
data_types = data_frame.dtypes
for columna, dtype in data_types.items():
    if dtype != 'object':
        variables_continuas.append(columna)

#Creando un diccionario que almacena los valores perdidos
valores_perdidos = {}
for variable in variables_continuas:
    valores_perdidos[variable] = None

#Agregando los valores perdidos a cada variable
for i, variable in enumerate(valores_perdidos):
    valor_perdido = data_frame[variable].isna()
    perdidos = valor_perdido.sum()
    valores_perdidos[variable] = perdidos

df_valores_perdidos = pd.DataFrame([valores_perdidos]).transpose()
df_valores_perdidos.index.name = "Variable"
print('Cantidad de datos perdidos para cada variable categorica')
df_valores_perdidos
```

```python
#Descriptivos de cada variable continua
#Creando un diccionario que almacena los estadisticos de cada variable
descriptivos = {}
for variable in variables_continuas:
    descriptivos[variable] = []

for i, variable in enumerate(variables_continuas):
    datos = data_frame[variable].describe().loc[['mean', 'std', 'max', 'min']]
    descriptivos[variable] = datos
    
df_desciptivos = pd.DataFrame(descriptivos).transpose()
df_desciptivos
```
