El conjunto de datos "diamonds" contiene contiene atributos físicos y precios de 53940 diamantes. A continuación, se describen las columnas o variables que se encuentran en el conjunto de datos:

*  **Carat:** Peso del diamante en quilates (una medida de peso para gemas, donde 1 quilate equivale a 0.2 gramos).

*  **Cut:** Calidad del corte del diamante, que puede tomar los siguientes valores: "Fair" (regular), "Good" (bueno), "Very Good" (muy bueno), "Premium" (premium) y "Ideal" (ideal).

*  **Color:** Color del diamante, que se codifica de la letra 'J' (peor) a la 'D' (mejor).

*  **Clarity:** Claridad del diamante, que describe la presencia de inclusiones (imperfecciones) en el diamante. Los valores van desde "I1" (incluido 1) hasta "IF" (sin inclusiones visibles).

*  **Depth:** Profundidad del diamante como un porcentaje de su diámetro total, que puede ayudar a determinar la calidad del corte.

*  **Table:** Ancho de la parte superior del diamante en relación con su punto más ancho.

*  **Price:** Precio del diamante en dólares estadounidenses.

*  **x, y, z:** Longitud, ancho y profundidad en milímetros del diamante.

En el mapa de calor de variables numéricas se observa que las variables con mayor correlación son *carat (peso del diamante en quilates)* y *precio* respecto a las dimensiones de los diamantes, sin embargo, las variables *depth (profundidad del diamante)* y *table (ancho de la parte superior)* tienen poca correlación con las demás variables.

```python
# Creación de la columna "Volumen" a partir de la multiplicación de las varaibles x(longitud), y(ancho) y profundidad(z):

diamantes_['volumen'] = diamantes_['x'] * diamantes_['y'] * diamantes_['z']

# Eliminación de las columnas x,y,z:

columnas_a_eliminar = ['x', 'y', 'z']
diamantes = diamantes_.drop(columns=columnas_a_eliminar)
diamantes
```

**a. (10%) Teniendo en cuenta los datos suministrados: ¿De qué variable o variables depende el valor de los diamantes?**

```python
# Modelo No. 1 con todas las variables categóricas y numéricas, con la variable "price" como dependiente:

dependent_var = 'price'
independent_vars = diamantes.columns.tolist()
independent_vars.remove(dependent_var)  
formula_1 = dependent_var + ' ~ ' + ' + '.join(independent_vars)

modelo1 = smf.ols(formula_1, data=diamantes).fit()

print(modelo1.summary())
```

Se observa que las variables continuas con mayor nivel de correlación respecto al precio de los diamantes son sus *kilates (carat)* y su *tamaño (table)*, dado por el volumen de los mismos. Al realizar un modelo de regresión lineal con la totalidad de las variables tanto numéricas como categoricas se observa que salvo las variables de las dimensiones *(x, y, z)*, todas en conjunto son significativas para explicar el precio de los diamantes. 

No obstante, al evaluar en regresión lineal simple cada variable, los predictores profundidad, color, corte, mesa y claridad presentan ajustes (R2) por debajo de 0,1.

```python
# Análisis de variable numérica "carat" y la variable "price":

modelo_rl = smf.ols('price ~ carat', data = diamantes).fit()

print(modelo_rl.summary())
```

Se observa que el modelo de regresión con la variable *carat* puede explicar aproximadamente un *84.9%* de la variabilidad del precio de los diamantes, quedando un *15.1%* de la variabilidad en los precios que podría estar influenciada por otras variables no incluidas en este análisis.

```python
# Análisis de variable numérica "depth" y la variable "price":

modelo_rl = smf.ols('price ~ depth', data = diamantes).fit()

print(modelo_rl.summary())
```

Bajo el modelo de regresión lineal con la variable *depth (profundidad)* como independiente, se puede concluir que esta no explica la variabilidad del comportamiento del precio de los diamantes.

```python
# Análisis de variable numérica "table" y la variable "price":

modelo_rl = smf.ols('price ~ table', data = diamantes).fit()

print(modelo_rl.summary())
```

Con este modelo se puede explicar alrededor del *1.6%* de las variaciones observadas en los precios de los diamantes en función de la variable *table (ancho de la parte superior)*, sin embargo, queda un *98.4%* de la variabilidad en los precios que no se ha tenido en cuenta en este modelo y que podría estar influenciada por otras variables no incluidas en el análisis.

```python
# Análisis de variable numérica "volumen" y la variable "price":

modelo_rl = smf.ols('price ~ volumen', data = diamantes).fit()

print(modelo_rl.summary())
```

A diferencia de los modelos anteriores, se observa que el modelo de regresión con la variable *volumen* puede explicar aproximadamente un *81.4%* de la variabilidad del precio de los diamantes, quedando un *18.6%* de la variabilidad en los precios que podría estar influenciada por otras variables no incluidas en este análisis.

Las variables *carat* y *volumen* al tener una fuerte correlación con la variable *precio* se podrian considerar las mas aptas como variables predictoras de esta ultima variable.

```python
# Modelo No.2 retirando las variables no significativas en modelo 1 (así la prueba F indicadra significancia)

# Modelo No.2 = smf.ols('price ~ volumen + table + depth + clarity + color + cut + carat', data=diamantes).fit()

modelo2 = smf.ols('np.log(price) ~ np.log(carat) + C(cut)', data=diamantes).fit()

print(modelo2.summary())
```

Al retirar las variables no significativas del modelo, se observa que se puede explicar el 93.7% de la variabilidad del comportamiento de la variable precio, quedando un *6.3%* de la variabilidad en los precios que podría estar influenciada por otras variables no incluidas en este análisis.

```python
#Análisis de residuos

# Calcular los residuos
residuos = modelo2.resid

# Calcular los valores estimados
valores_ajustados = modelo2.fittedvalues

# gráfico de dispersión de los residuos frente a los valores ajustados
plt.figure(figsize=(8, 6))
plt.scatter(valores_ajustados, residuos)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Valores Ajustados')
plt.ylabel('Residuos')
plt.title('Gráfico de Residuos vs. Valores Ajustados')

# Mostrar el gráfico
plt.show()
```

Se observa que el patrón de los residuos tiene un comportamiento inversamente propocional y se agrupan entre el intervalo de 1 a -1. Se puede concluir que el modelo ajustado es adecuado para predecir el valor de los precios de los diamantes.

```python
# Análisis de correlación
residuos = modelo2.resid

# Crear un QQ-Plot de los residuos
sm.qqplot(residuos, line='s', fit=True, alpha=0.5)
plt.title('QQ-Plot de los Residuos')
plt.show()
```

Se observa que los residuos caen a lo largo de la linea recta, entonces los residuos se distribuyen aproximadamente normal y por ende se puede concluir que el modelo de regresión es bastante robusto respecto a las desviaciones de la normalidad.

Ecuación del modelo de regresión lineal

$\widehat{y}= +\beta _{1}*X + \beta _{2}*Y + \beta _{3}*Z + \epsilon$

```python
# Definir las variables predictoras (X) y la variable objetivo (y)

X_tam = diamantes_[['x', 'y','z']]
y_tam = diamantes_['price']

modelo_tam = LinearRegression().fit(X_tam,y_tam)

B0 = modelo_tam.intercept_
B1,B2,B3 = modelo_tam.coef_

print(f'Intercepto (B0): {B0}')
print(f'Intercepto (B1): {B1}')
print(f'Intercepto (B2): {B2}')
print(f'Intercepto (B3): {B3}')
```

Al obtener un intercepto $\beta _{0}$ con un signo negativo se puede concluir que el precio de un diamante es cero cuando las variables independientes sean iguales a cero, sin embargo, para este caso no es posible este escenario dado que no existiria el objeto como tal.

```python
precios =[]

for i in range(len(x)):

    precio = B0 + B1*x[i] + B2*y[i] + B3*z[i]
    precios.append(precio)

precios
```

```python
# Al retirar el intercepto negativo del modelo:

precios =[]

for i in range(len(x)):

    precio = B1*x[i] + B2*y[i] + B3*z[i]
    precios.append(precio)

precios
```

* Se procede a crear la variable volumen la cual contiene la multiplicación de x,y,z y también a retirar el intercepto para no contar con valores negativos para el precio del diamante cuando no se presente su volumen, lo cual es equivalente a la inexistencia del diamante. 
De esta forma, el precio de un diamante está dado por  34.4754 veces su volumen: Precio = 34,4754Volumen, por lo cual al aumentar en un centímetro cúbico el tamaño de un diamante, su precio aumentará 34,47 USD.

Ecuación del modelo de regresión lineal simple

```python
# Implementa tu respuesta en esta celda

# Crear un modelo de regresión lineal con ambas variables predictoras

modelo_rl = smf.ols('price ~ carat', data = diamantes).fit()

# Impresión de resultados

print(modelo_rl.summary())
```

Dado que el modelo presenta un R2 del 84,9%, y que los resultados muestran que la variable quilates es significativa para explicar el precio de los diamantes, se rechaza la hípotesis que los quiltes no influyen en el valor de los diamantes.

```python
# Implementa tu respuesta en esta celda

# Estimación de la duración en minutos

archivo = "Archivos/taxis.csv" 
df2 = pd.read_csv(archivo)
df2['inicio'] = pd.to_datetime(df2['pickup'])
df2['fin'] = pd.to_datetime(df2['dropoff'])
df2['duracion'] = (df2['fin'] - df2['inicio']).dt.total_seconds() / 60

#df2['minutos'] = pd.to_numeric(df2['duracion'])
df2.head()
```

```python
# Implementa tu respuesta en esta celda
# Modelo de regresión lineal simple
# y = Total del valor de la carrera
# X = Duración de la carrera(minutos)

modelo_2a = smf.ols('total ~ duracion', data = df2).fit()
print(modelo_2a.summary())
```

```python
#Coeficiente de correlación:

correlacion_2a = df2['duracion'].corr(df2['total'])

print(f"Coeficiente de correlación: {correlacion_2a}")
```

```python
# Implementa tu respuesta en esta celda

# Análisis de residuos

residuos2a = modelo_2a.resid

# Calcular los valores estimados
valores_ajustados = modelo_2a.fittedvalues

# gráfico de dispersión de los residuos frente a los valores ajustados

#plt.figure(figsize=(8, 6))
plt.scatter(valores_ajustados, residuos2a)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Valores Ajustados')
plt.ylabel('Residuos')
plt.title('Gráfico de Residuos vs. Valores Ajustados ( Total ~ duración )')

# Crear un QQ-Plot de los residuos
sm.qqplot(residuos2a, line='s', fit=True, alpha=0.5)
plt.title('QQ-Plot de los Residuos ( Total ~ duración )')
plt.show()

# Histograma de los residuos
plt.hist(residuos2a, bins=70, color='blue', alpha=0.7)
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')
plt.title('Histograma de Residuos ( Total ~ duración )')
plt.show()
# Mostrar el gráfico
plt.show()

# Prueba de homocedasticidad
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

# Prueba Breusch Pagan de homocedasticidad
bp_test = het_breuschpagan(modelo_2a.resid, modelo_2a.model.exog)
print(f"Breusch-Pagan test :\nLM Statistic: {bp_test[0]}\nLM-Test p-value: {bp_test[1]}\nF-Statistic: {bp_test[2]}\nF-Test p-value: {bp_test[3]}\n")
```

```python
# Implementa tu respuesta en esta celda
df5 = pd.read_csv('Archivos/dataset_boston.csv', encoding = "UTF-8",sep= ";")
df5.head()
```

```python
# Implementa tu respuesta en esta celda
dependent_var = 'MDEV'
independientes = df5.columns.tolist()
independientes.remove(dependent_var)  # remove the dependent variable from the list
formula = dependent_var + ' ~ ' + ' + '.join(independientes)

model5 = smf.ols(formula, data=df5).fit()
print(model5.summary())
```

Se genero un modelo de regresiónlineal donde nuestra variable dependiente es el valor medio de casas ocupadas por sus dueños ***(MDEV)*** y las demas variables como predictoras. Un análisis inicial permite identificar variables que no son estadisticamente significativas por la prueba P>|t| como ***'INDUS***' y ***'AGE'***.

Para analizar la correlación entre las variables predictoras y la variable de salida, acudimos al gráfico de calor el cual me permite visualizar todas las correlaciones.

```python
import seaborn as sns

correlation_matrix = df5.corr() 
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Corrlación modelo de estimación de precios')
plt.show()

correlation_values = correlation_matrix.loc['MDEV']
print(correlation_values)
```

Analizando el mapa de calor, primero evaluamos que las variables que tienen un coeficiente negativo influyen de manera específica sobre la variable independiente. Estas variables están relacionadas con factores socio-económicos y de ubicación de la vivienda; estas variables fueron agregadas en su totalidad al modelo del literal b.

**Variables con coeficiente negativo:** CRIM, INDUS, NOX, AGE, RAD, TAX, PTRATIO, LSTAT

Por último, la variable asociada a la cantidad de habitaciones (RM) fue agregada bajo el supuesto que la cantidad de espacios dedicados tienen una fuerte relación directa con el precio de la vivienda. En el siguiente literal, se podrá observar el modelo con las variables descritas en este parrafo.

b. (25%) Con el conjunto de datos suministrado de Boston, realizen un modelo de regresion múltiple que prediga el precio medio de las viviendas en diferentes áreas en función de diversas características. Si considera necesario experimentar con diferentes configuraciones adelante.

```python
columnas_modelo=['RM','LSTAT','TAX','PTRATIO','RAD','AGE','NOX','INDUS','CRIM','MDEV']

df5=df5[columnas_modelo]

dependent_var = 'MDEV'
independientes = df5.columns.tolist()
independientes.remove(dependent_var)  # remove the dependent variable from the list
formula = dependent_var + ' ~ ' + ' + '.join(independientes)

model5 = smf.ols(formula, data=df5).fit()
print(model5.summary())
```

El modelo obtenido muestra un indice de bondad de ajuste del (R^2 adj) del 0.693, indicando el que 69.3% de la variabilidad de la variable dependiente, depende de la variables predictoras. 

***Prueba hipotesis para significancia global del modelo***

* $H_{0}$: Los coeficientes de la regresión son iguales a cero. Indica que las variables predictoras no tendrían un efecto significativo en el la variable de salida.
* $H_{1}$: Al menos uno de los coeficientes de regresión es diferente de cero. En este caso, al menos una variable tiene un efecto significativo en en la variable de salida.

Con un F-estadistico de 127.7 y un p-value aprox. de cero. Con un intervalo de confianza del 95%, el p valor es menor y el valor F es mayor que el valor crítico. Permite rechazar la hipotensis HO, concluyendo que al menos una de las variables tiene un efecto significativo en la variable de salida del modelo.

***Prueba hipotesis para coeficientes individuales***

* $H_{0}$: El coeficiente es igual a cero.
* $H_{1}$: El coeficiente es diferente de cero.

Visualizando los p-values asociados a cada coeficiente y asumiendo un a=0.05, concluimos que podemos rechazar $H_{0}$ y concluir que cada coeficiente es diferente de cero, estableciendo que cada variable predictora tiene un efecto significativo en la salida.

***Comprobación de supuestos***

```python
residuos5=model5.resid
valores_ajustados5=model5.fittedvalues

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].scatter(valores_ajustados5, residuos5, alpha=0.5)
axes[0].set_xlabel('adjusted values')
axes[0].set_ylabel('Residuals')
axes[0].set_title('Residuals graph')
axes[0].grid(True)
axes[0].axhline(0, color='red', linestyle='--')

axes[1].hist(residuos5, bins=20, color='darkorange', edgecolor='dodgerblue')
axes[1].set_xlabel('Residuals')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Histogram of residuals')
axes[1].grid(True)

plt.show()
```

El gráfico de residuos no presenta una distribución constante alrededor la linea centrada  en cero. En este caso, la media de los residuos es diferente de cero, ***violando el principio de varianza constante.***

En este caso, se realiza la prueba de Breush-Pagan para determinar la homoscedasticidad del modelo. 

* $H_{0}$: **Homoscedasticidad** - Errores con varianza constante.
* $H_{1}$: **Heteroscedasticidad** - La varianza de los errores no es constante.

```python
bp_test = het_breuschpagan(model5.resid, model5.model.exog)
print(f"Breusch-Pagan test :\nLM Statistic: {bp_test[0]}\nLM-Test p-value: {bp_test[1]}\nF-Statistic: {bp_test[2]}\nF-Test p-value: {bp_test[3]}\n")
```

En este caso, la prueba es significativa, permitiendo rechazar $H_{0}$. Los residuos de la regresion presentan heroscedasticidad.

```python
import seaborn as sns
df5['predicc'] = model5.fittedvalues

sns.lmplot(x=dependent_var, y='predicc', data=df5, ci=None, scatter_kws={'color': 'skyblue'}, line_kws={'color': 'red'})

plt.xlabel('Valores Observados')
plt.ylabel('Valores Predichos')
plt.title('Gráfico de Dispersión con Recta de Regresión')
plt.grid()
plt.show()
```

```python
sm.qqplot(residuos5, line='s', color='darkgrey', marker='o')
plt.title('Normal Q-Q Plot, fit_1')
plt.gca().get_lines()[1].set_color('dodgerblue')
plt.gca().get_lines()[1].set_linewidth(2)
plt.show()
```

En este caso, el QQ plot ilustra valores extremos los cuales se encuentran apartados de la linea diagonal. En este caso, nuestro modelo presenta valores atípicos en su lado derecho, efecto visible en el histograma de residuos donde presenta una asimetría hacia la izquierda, donde la mayor concentración de residuos se encuentran a la derecha.

#### Transformación

Debido a que no se cumplen lo supuestos de linealidad, se propone realizar una transformación para disminuir la variabilidad. Para ello se propone realizar un logaritmo a la variable dependiente y generar polinomios cuadrados para las variales **RM** y **LSTAT**. 

Se obtuvieron los siguientes resultados:

```python
import statsmodels.formula.api as smf

model6 = smf.ols('np.log(MDEV) ~ RM+LSTAT+TAX+PTRATIO+RAD+AGE+NOX+INDUS+CRIM+I(RM**2)+I(LSTAT**2)',data=df5).fit()
print(model6.summary())
```

En este caso, el logaritmo en base 10 del precio es estimado por las variables socio-económicas y de ubicación mas la cantidad de habitaciones elevada al cuadrado mas el porcentaje de personas pobres elevadas al cuadrado.
Se puede observar como e Adj. R-squared mejora de 0.693 a 0.797, lo que indica que el modelo se ajusta mejor en la transformación y todas las variables predictoras son estadisticamente significativas.

La ecuación para la predicción del precio de las casas en Boston estaría dado por:

$$\hat Y = 6.72 - 0.79RM - 0.05LSTAT - 0.001TAX - 0.03PTRATIO + 0.01RAD + 0.001AGE - 0.55NOX + 0.01INDUS - -0-1CRIM + 0.07RM^2 + 0.001LSTAT^2$$


Al repetir la pruebas, el modelo cumple con supuestos de significancia total e individual al 95% de confianza.

```python
import seaborn as sns
df5['predicc2'] = model6.fittedvalues

sns.lmplot(x=dependent_var, y='predicc2', data=df5, ci=None, scatter_kws={'color': 'skyblue'}, line_kws={'color': 'red'})

plt.xlabel('Valores Observados')
plt.ylabel('Valores Predichos')
plt.title('Gráfico de Dispersión con Recta de Regresión')
plt.grid()
plt.show()
```

Se puede observar que a comparación con el gráfico anterior, los puntos están un poco mas cerca a la recta de regresión. Los valores obsrvados entre 0 a 30 tambien presentan una menor variabilidad y tienen un indicio de linealidad.

```python
residuos6=model6.resid
valores_ajustados6=model6.fittedvalues

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].scatter(valores_ajustados6, residuos6, alpha=0.5)
axes[0].set_xlabel('adjusted values')
axes[0].set_ylabel('Residuals')
axes[0].set_title('Residuals graph')
axes[0].grid(True)
axes[0].axhline(0, color='red', linestyle='--')

axes[1].hist(residuos6, bins=20, color='darkorange', edgecolor='dodgerblue')
axes[1].set_xlabel('Residuals')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Histogram of residuals')
axes[1].grid(True)

plt.show()
```

```python
sm.qqplot(residuos6, line='s', color='darkgrey', marker='o')
plt.title('Normal Q-Q Plot, fit_1')
plt.gca().get_lines()[1].set_color('dodgerblue')
plt.gca().get_lines()[1].set_linewidth(2)
plt.show()
```

Al analizar el gráfico de dispersión de residuos, se puede analizar una mejora en la distribución de estos, respecto al anterior. Antes el gráfico de residuos presentaba un comportamiento cuadriatico con alta variabilidad, ahora ese comportamiento es mas líneal. Asi mismo el histograma permite ver que los datos se estan estandarizando y se dejan de lado la cola derecha que genereaba una asimetría positiva.

Igualmente, cuando se analiza el Q-Q plot, se puede ver como los datos estan mas cerca de la diagonal y las colas se han disminuido significativamente.

Ahora se procede a realizar la prueba de Breusch-Pagan.

```python
bp_test = het_breuschpagan(model6.resid, model6.model.exog)
print(f"Breusch-Pagan test :\nLM Statistic: {bp_test[0]}\nLM-Test p-value: {bp_test[1]}\nF-Statistic: {bp_test[2]}\nF-Test p-value: {bp_test[3]}\n")
```

Al repetir la prueba de Breusch-Pagan, observamos que aun se presenta heteroscedasticidad en los residuos del modelo.
