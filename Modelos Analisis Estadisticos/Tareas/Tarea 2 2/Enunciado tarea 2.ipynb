```python
# Implementa tu respuesta en esta celda
ruta3='C://Users//guavitaj//Documents//semestre1//modelos estadisticos//tareas//Taller 2//Taller 2//Archivos//dataset_boston.csv'
df5 = pd.read_csv(ruta3,sep=';')
df5.head()

```

```python
# Implementa tu respuesta en esta celda
dependent_var = 'MDEV'
independientes = df5.columns.tolist()
independientes.remove(dependent_var)  # remove the dependent variable from the list
formula = dependent_var + ' ~ ' + ' + '.join(independientes)

model5 = smf.ols(formula, data=df5).fit()
print(model5.summary())
```

Se genero un modelo de regresiónlineal donde nuestra variable dependiente es el valor medio de casas ocupadas por sus dueños ***(MDEV)*** y las demas variables como predictoras. Un análisis inicial permite identificar variables que no son estadisticamente significativas por la prueba P>|t| como ***'INDUS***' y ***'AGE'***.

Para analizar la correlación entre las variables predictoras y la variable de salida, acudimos al gráfico de calor el cual me permite visualizar todas las correlaciones.

```python
import seaborn as sns

correlation_matrix = df5.corr() 
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Corrlación modelo de estimación de precios')
plt.show()

correlation_values = correlation_matrix.loc['MDEV']
print(correlation_values)
```

Analizando el mapa de calor, primero evaluamos que las variables que tienen un coeficiente negativo influyen de manera específica sobre la variable independiente. Estas variables están relacionadas con factores socio-económicos y de ubicación de la vivienda; estas variables fueron agregadas en su totalidad al modelo del literal b.

**Variables con coeficiente negativo:** CRIM, INDUS, NOX, AGE, RAD, TAX, PTRATIO, LSTAT

Por último, la variable asociada a la cantidad de habitaciones (RM) fue agregada bajo el supuesto que la cantidad de espacios dedicados tienen una fuerte relación directa con el precio de la vivienda. En el siguiente literal, se podrá observar el modelo con las variables descritas en este parrafo.

b. (25%) Con el conjunto de datos suministrado de Boston, realizen un modelo de regresion múltiple que prediga el precio medio de las viviendas en diferentes áreas en función de diversas características. Si considera necesario experimentar con diferentes configuraciones adelante.

```python
columnas_modelo=['RM','LSTAT','TAX','PTRATIO','RAD','AGE','NOX','INDUS','CRIM','MDEV']

df5=df5[columnas_modelo]

dependent_var = 'MDEV'
independientes = df5.columns.tolist()
independientes.remove(dependent_var)  # remove the dependent variable from the list
formula = dependent_var + ' ~ ' + ' + '.join(independientes)

model5 = smf.ols(formula, data=df5).fit()
print(model5.summary())
```

El modelo obtenido muestra un indice de bondad de ajuste del (R^2 adj) del 0.693, indicando el que 69.3% de la variabilidad de la variable dependiente, depende de la variables predictoras negativas y RM. 

***Prueba hipotesis para significancia global del modelo***

* $H_{0}$: Los coeficientes de la regresión son iguales a cero. Indica que las variables predictoras no tendrían un efecto significativo en el la variable de salida.
* $H_{1}$: Al menos uno de los coeficientes de regresión es diferente de cero. En este caso, al menos una variable tiene un efecto significativo en en la variable de salida.

Con un F-estadistico de 127.7 y un p-value aprox. de cero. Con un intervalo de confianza del 95%, el p valor es menor y el valor F es mayor que el valor crítico. Permite rechazar la hipotensis HO, concluyendo que al menos una de las variables tiene un efecto significativo en la variable de salida del modelo.

***Prueba hipotesis para coeficientes individuales***

* $H_{0}$: El coeficiente es igual a cero.
* $H_{1}$: El coeficiente es diferente de cero.

Visualizando los p-values asociados a cada coeficiente y asumiendo un a=0.05, concluimos que podemos rechazar $H_{0}$ y concluir que cada coeficiente es diferente de cero, estableciendo que cada variable predictora tiene un efecto significativo en la salida.

***Comprobación de supuestos***

```python
residuos5=model5.resid
valores_ajustados5=model5.fittedvalues

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].scatter(valores_ajustados5, residuos5, alpha=0.5)
axes[0].set_xlabel('adjusted values')
axes[0].set_ylabel('Residuals')
axes[0].set_title('Residuals graph')
axes[0].grid(True)
axes[0].axhline(0, color='red', linestyle='--')

axes[1].hist(residuos5, bins=20, color='darkorange', edgecolor='dodgerblue')
axes[1].set_xlabel('Residuals')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Histogram of residuals')
axes[1].grid(True)

plt.show()
```

El gráfico de residuos no presenta una distribución constante alrededor la linea centrada  en cero. En este caso, la media de los residuos es diferente de cero, ***violando el principio de varianza constante.***

En este caso, se realiza la prueba de Breush-Pagan para determinar la homoscedasticidad del modelo. 

* $H_{0}$: **Homoscedasticidad** - Errores con varianza constante.
* $H_{1}$: **Heteroscedasticidad** - La varianza de los errores no es constante.

```python
bp_test = het_breuschpagan(model5.resid, model5.model.exog)
print(f"Breusch-Pagan test :\nLM Statistic: {bp_test[0]}\nLM-Test p-value: {bp_test[1]}\nF-Statistic: {bp_test[2]}\nF-Test p-value: {bp_test[3]}\n")
```

En este caso, la prueba es significativa, permitiendo rechazar $H_{0}$. Los residuos de la regresion presentan heroscedasticidad.

```python
import seaborn as sns
df5['predicc'] = model5.fittedvalues


sns.lmplot(x=dependent_var, y='predicc', data=df5, ci=None, scatter_kws={'color': 'skyblue'}, line_kws={'color': 'red'})

plt.xlabel('Valores Observados')
plt.ylabel('Valores Predichos')
plt.title('Gráfico de Dispersión con Recta de Regresión')
plt.grid()
plt.show()
```

En este caso, el QQ plot ilustra valores extremos los cuales se encuentran apartados de la linea diagonal. #### Transformación

Debido a que no se cumplen lo supuestos de linealidad, se propone realizar una transformación para disminuir la variabilidad. Para ello se propone realizar un logaritmo a la variable dependiente y generar polinomios cuadrados para las variales **RM** y **LSTAT**. 

Se obtuvieron los siguientes resultados:

```python
import statsmodels.formula.api as smf

model6 = smf.ols('np.log(MDEV) ~ RM+LSTAT+TAX+PTRATIO+RAD+AGE+NOX+INDUS+CRIM+I(RM**2)+I(LSTAT**2)',data=df5).fit()
print(model6.summary())
```

En este caso, el logaritmo en base 10 del precio es estimado por las variables socio-económicas y de ubicación mas la cantidad de habitaciones elevada al cuadrado mas el porcentaje de personas pobres elevadas al cuadrado.
Se puede observar como e Adj. R-squared mejora de 0.693 a 0.797, lo que indica que el modelo se ajusta mejor en la transformación y todas las variables predictoras son estadisticamente significativas.

La ecuación para la predicción del precio de las casas en Boston estaría dado por:
$$\hat Y = 6.72 - 0.79RM - 0.05LSTAT - 0.001TAX - 0.03PTRATIO + 0.01RAD + 0.001AGE - 0.55NOX + 0.01INDUS - -0-1CRIM + 0.07RM^2 + 0.001LSTAT^2$$

Al repetir la pruebas, el modelo cumple con supuestos de significancia total e individual al 95% de confianza.

```python
import seaborn as sns
df5['predicc2'] = model6.fittedvalues

sns.lmplot(x=dependent_var, y='predicc2', data=df5, ci=None, scatter_kws={'color': 'skyblue'}, line_kws={'color': 'red'})

plt.xlabel('Valores Observados')
plt.ylabel('Valores Predichos')
plt.title('Gráfico de Dispersión con Recta de Regresión')
plt.grid()
plt.show()
```

Se puede observar que a comparación con el gráfico anterior, los puntos están un poco mas cerca a la recta de regresión. Los valores obsrvados entre 0 a 30 tambien presentan una menor variabilidad y tienen un indicio de linealidad.
