```python
# Write your code here.
import pandas as pd
csv = "/Users/santiagoromero/Documents/Maestria/Modelos Analisis Estadisticos/penguins_classification.csv"
data_frame = pd.read_csv(csv)
data_frame.head()
```

¬øCu√°les son las diferentes especies de ping√ºinos disponibles en el conjunto de datos y cu√°ntas muestras de cada especie hay? Sugerencia: seleccione la columna de la derecha y utilice el m√©todo [`value_counts`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html).

```python
# Write your code here.
data_frame['Species'].value_counts()
```

Trazar histogramas para las caracter√≠sticas num√©ricas

```python
# Write your code here.
import matplotlib.pyplot as plt
plt.hist(data_frame['Culmen Length (mm)'], edgecolor='black', bins=10)
plt.title("Culmen Length for all penguin species")
plt.xlabel("Culmen Length (mm")
plt.ylabel("Cantidad de ping√ºinos")
plt.show()

plt.hist(data_frame['Culmen Depth (mm)'], color='red', edgecolor='black', bins=10)
plt.title("Culmen Depth for all penguin species")
plt.xlabel("Culmen Depth (mm)")
plt.ylabel("Cantidad de ping√ºinos")
plt.show()
```

Muestre la distribuci√≥n de caracter√≠sticas para cada clase. Sugerencia: utilice [`seaborn.pairplot`](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

```python
# Write your code here.
import seaborn as sns
sns.pairplot(data_frame, hue="Species")
```
