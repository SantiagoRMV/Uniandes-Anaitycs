### **Parte Teórica**:

1. **¿Qué son los residuos en una regresión lineal y por qué son importantes?**


Son las diferencias entre los valores observados y los estimados por el modelo. Son importantes por que permite:
1. Evaluar que tan bien se ajusta el modelo a los datos observados. Si son pequeños y aleatorios, significa que el modelo es adecuado para describir la relación de las variables
2. Evaluar patrones de aumento o disminución del error (homocedasticos o heterocedastico)
3. Evaluar las supociciones de regresión lineal 
4. Identificación de datos atipicos


2. **Lista y describe las principales suposiciones en un modelo de regresión lineal.**  



1. Linealidad: La relación entre la variable independiente y dependiente es lineal
2. Indeprendencia de errores: Se asume que son independientes entre si y no estan correlacionados con otros errores
3. Homoscedasticidad: Los errores no tienen varianza 
4. Normalidad: Los errores siguen una distribución normal con media 0 


3. **¿Cómo se pueden identificar patrones en los residuos y qué indican estos patrones?**


Los patrones se pueden identificar como:
* Homoscedasticidad: La varianza de los errores es constante (lo que se espera de un modelo que se ajuste)
* Heteroscedasticidad: La varianza de los errores no es constante


4. **¿Cuál es el propósito de una transformación de datos en regresión?** 


Se aplica transformación de datos a las respuestas y los predictores para corregir problemas con el diagnostico de modelos y hacer que se ajuste mejor a los datos

Problemas: No normalidad, multicolinalidad, varianza no homogenea, no linealidad


### **Parte Práctica**:

**Objetivo:** Realizar un análisis de residuos en un modelo de regresión lineal y sugerir mejoras basadas en el análisis.


```python
import seaborn as sns
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from scipy import stats

diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
vacios = diamonds.isna().any().any()
print(vacios)
```


**Instrucciones:**



1. **Ajustar un modelo de regresión lineal.**  



   Utilice un conjunto de datos de su elección y ajuste un modelo de regresión lineal inicial.


```python
diamonds.describe()
```

```python
#EDA
sns.set(style='ticks', color_codes=True)
sns.pairplot(diamonds, kind='reg')
plt.show()
```

```python
#Matriz de correlación
corr = diamonds.corr()
sns.heatmap(corr, xticklabels= corr.columns.values, yticklabels= corr.columns.values)
plt.show()
```

Las variables con mayor correlación son Depth y Table

```python
#Modelo de regresión lineal
model = smf.ols('price ~ z', data = diamonds).fit()
model.params
```

```python
print(model.summary())
```


2. **Graficar los residuos.**  
   - Contra los valores predichos: Esta gráfica ayuda a identificar problemas de homocedasticidad.
   - Contra las variables independientes: Esto puede mostrar si hay relaciones no lineales.
   - Histograma de residuos: Para verificar la suposición de normalidad.


Valores predichos / Variables independientes

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))

# Fitted vs Residuals Plot
ax[0].scatter(model.fittedvalues, model.resid, color='grey', s=60)
ax[0].axhline(y=0, color='darkorange', linewidth=2)
ax[0].set_title('Fitted versus Residuals')
ax[0].set_xlabel('Fitted')
ax[0].set_ylabel('Residuals')

# Q-Q Plot using scipy's probplot
stats.probplot(model.resid, plot=ax[1], fit=True)
ax[1].get_lines()[0].set_markerfacecolor('darkgrey')  # Changing the color of the dots to 'darkgrey'
ax[1].get_lines()[1].set_color('dodgerblue')  # Changing the color of the line to 'dodgerblue'
ax[1].set_title('Normal Q-Q Plot')
```

Histograma de residuos

```python
def plot_residual_histogram(model, title):
    residuals = model.resid
    plt.hist(residuals, bins=20, color='darkorange', edgecolor='dodgerblue')
    plt.xlabel('Residuals')
    plt.ylabel('Frequency')
    plt.title(title)

# Plotting
plt.figure(figsize=(15,5))
plot_residual_histogram(model, 'Histogram of Residuals model')
plt.tight_layout()
plt.show()
```


3. **Identificar patrones.**  
   Basándose en las gráficas, identifique si hay algún patrón en los residuos. 


Se pueden ver patrones de linealidad en los residuos, sin embargo tambien cuentan con heteroscedasticidad, por lo que e necesario realiar una transformación para ajustar los datos


4. **Realizar transformaciones (si es necesario).**  
   Si se identifican patrones, considere transformaciones como logaritmos, raíces cuadradas o transformaciones Box-Cox.


```python
model_log = smf.ols('np.log(price) ~ z', data=diamonds).fit()
```

```python
# Plotting
plt.scatter(diamonds['z'], np.log(diamonds['price']), color='grey', s=60)
plt.plot(diamonds['z'], model_log.fittedvalues, color='darkorange', linewidth=2)
plt.title('')
plt.xlabel('z')
plt.ylabel('Log(price)')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))

# Fitted vs Residuals Plot
ax[0].scatter(model_log.fittedvalues, model_log.resid, color='grey', s=60)
ax[0].axhline(y=0, color='darkorange', linewidth=2)
ax[0].set_title('Fitted versus Residuals')
ax[0].set_xlabel('Fitted')
ax[0].set_ylabel('Residuals')

# Q-Q Plot using scipy's probplot
stats.probplot(model_log.resid, plot=ax[1], fit=True)
ax[1].get_lines()[0].set_markerfacecolor('darkgrey')  # Changing the color of the dots to 'darkgrey'
ax[1].get_lines()[1].set_color('dodgerblue')  # Changing the color of the line to 'dodgerblue'
ax[1].set_title('Normal Q-Q Plot')
```

```python
mark_mod_poly3 = smf.ols('price ~ z + I(z ** 2) + I(z ** 3)', data=diamonds).fit()
print(mark_mod_poly3.summary())
```

```python
# Polynomial Model Visualization
xplot = np.linspace(0, 16, 100)
plt.scatter(diamonds['z'], diamonds['price'], s=40)
plt.plot(xplot, mark_mod_poly3.predict(pd.DataFrame({'z': xplot})), color='red', linestyle='dotted')
plt.xlabel('Z')
plt.ylabel('Price')
plt.show()
```


5. **Reajustar y comparar.**  
   Ajuste nuevamente el modelo con las variables transformadas y compare los residuos del nuevo modelo con el anterior.



```python
print(model_log.summary())
```

```python
rmse_manual_original = np.sqrt(np.mean((diamonds['price'] - model.fittedvalues) ** 2))
rmse_manual_log_transformed = np.sqrt(np.mean((diamonds['price'] - np.exp(model_log.fittedvalues)) ** 2))

print('Manual RMSE Original:', rmse_manual_original)
print('Manual RMSE Log Transformed:', rmse_manual_log_transformed)
```
